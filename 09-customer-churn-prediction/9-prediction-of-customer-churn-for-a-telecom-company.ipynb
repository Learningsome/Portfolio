# Модель предсказания оттока клиентов

**Описание проекта**

Для оператор связи нужно сделать модель, которая будет прогнозировать отток клиентов. Если выяснится, что пользователь планирует уйти, ему будут предложены промокоды и специальные условия. Команда оператора собрала персональные данные о некоторых клиентах, информацию об их тарифах и договорах.
\
\
\
**Описание услуг**

Оператор предоставляет два основных типа услуг:
1. Стационарную телефонную связь. Возможно подключение телефонного аппарата к нескольким линиям одновременно.
2. Интернет. Подключение может быть двух типов: через телефонную линию (DSL, от англ. *digital subscriber line*, «цифровая абонентская линия») или оптоволоконный кабель (*Fiber optic*).


Также доступны такие услуги:
- Интернет-безопасность: антивирус (DeviceProtection) и блокировка небезопасных сайтов (OnlineSecurity);
- Выделенная линия технической поддержки (TechSupport);
- Облачное хранилище файлов для резервного копирования данных (OnlineBackup);
- Стриминговое телевидение (StreamingTV) и каталог фильмов (StreamingMovies).


За услуги клиенты могут платить каждый месяц или заключить договор на 1–2 года. Доступны различные способы расчёта и возможность получения электронного чека.


**Описание данных**

Данные состоят из файлов, полученных из разных источников:
- `personal_new.csv` - персональные данные клиента
    - gender - пол
    - SeniorCitizen - наличие пенсионного статуса по возрасту
    - Partner - наличие супруга/супруги
    - Dependents - наличие иждивенцев
- `contract_new.csv` - информация о договоре
    - BeginDate - дата начала пользования услугами
    - EndDate - дата окончания пользования услугами
    - Type - тип договора: ежемесячный, годовой и т.д
    - PaperlessBilling - выставления счёта по электронной почте
    - PaymentMethod - способ оплаты
    - MonthlyCharges - ежемесячные траты на услуги
    - TotalCharges - всего потрачено денег на услуги
- `internet_new.csv` - информация об интернет-услугах
    - InternetService - наличие услуг Интернет
    - OnlineSecurity - межсетевой экран
    - OnlineBackup - облачное хранилище файлов для резервного копирования данных
    - DeviceProtection - антивирус
    - TechSupport - выделенная линия технической поддержки
    - StreamingTV - онлайн-ТВ
    - StreamingMovies - онлайн-кинотеатр
- `phone_new.csv` - информация об услугах телефонии
    - MultipleLines - возможность подключения телефонного аппарата к нескольким линиям одновременно

Во всех файлах столбец `customerID` содержит код клиента.

Информация о договорах актуальна на 1 февраля 2020.

Объявим нужные функции

````python
# общая информация о df
def get_info(df: pd.DataFrame):
    """Функция вызывает подряд три основных метода head(), info() и describe(include='all')

    Args:
        df (pd.DataFrame): DataFrame, для которого нужно вывести общую информацию
    """
    display(df.head())
    print("-" * 50)
    df.info()
    print("-" * 50)
    display(df.describe(include="all"))


# CamelCase to snake_case
def camel_to_snake(row: str) -> str:
    """Функция переводит строку из CamelCase в snake_case

    Args:
        row (str): Строка, которую нужно изменить

    Returns:
        str: Измененная строка
    """
    name = re.sub(r"(?<!^)(?=[A-Z])", "_", row).lower()

    return name


# переименовать столбцы df
def rename_columns(df: pd.DataFrame):
    """Функция переводит названия столбцов DataFrame из CamelCase в snake_case

    Args:
        df (pd.DataFrame): DataFrame, в котором нужно изменить названия столбцов
    """
    try:
        df.columns = map(lambda column: camel_to_snake(column), df.columns)
        df.rename(
            {"customer_i_d": "customer_id", "streaming_t_v": "streaming_tv"},
            axis=1,
            inplace=True,
        )
        display(df.columns)
    except:
        display(df.columns)
        
        
def plot_histplot(df: pd.DataFrame, x: str, palette: str, title: str):
    """Функция рисует histplot

    Args:
        df (pd.DataFrame): DataFrame с данными для графика
        
        x (str): Признак, который необходимо изобразить на графике
        
        palette (str): Палетка seaborn
        
        title (str): Заголовок графика
    """
    sns.histplot(data=df, x=x, hue="churned", palette=palette)
    plt.ylabel('Количество записей')
    plt.title(title, size=20);
````

Введем дополнительный признак - `days_since_begin_date`, который будет показывать разницу между 'end_date' и 'begin_date'. 

````python
data_merged.end_date.head()
````

Для начала, вместо 'No' вставим дату актуальности датасета, таким образом, можно будет привести столбец к нужному типу данных, а, затем, и посчитать разницу между значениями.

### Кодирование и масштабирование признаков

Делим data_merged на train и test выборки, учитываем при этом необходимые критерии:
1. Отношение train к test - 3 к 1
2. shuffle=True
3. stratify=y т.к. сохраняем дисбаланс

````python
# разделение data_merged на признаки и таргет
X = data_merged.drop('churned', axis=1)
y = data_merged['churned']


# разделение X, y на train и test подвыборки
X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.25, shuffle=True, stratify=y, random_state=RANDOM_STATE
)
````

На трейн и тест разделили, теперь перейдем к кодированию.

Первая на очереди модель - LogisticRegression, для того, чтобы она работала корректно, необходимо трансформировать данные:

1. StandardScaler
2. OneHotEncoder

````python
#категориальные признаки для OHE
ohe_features = X_train.select_dtypes(include='object').columns.to_list()

# числовые признаки для скелера
num_features = ['monthly_charges', 'days_since_begin_date']

# подвыборки для lr
X_train_lr = X_train.copy()
X_test_lr = X_test.copy()
````

````python
# объявляем эстиматор
lr_model = LogisticRegression(class_weight='balanced', random_state=RANDOM_STATE)

# Создаем трансформер
col_transformer_lr = ColumnTransformer(
    transformers=[
        ('num_robust', RobustScaler(), num_features),
        ('ohe', OneHotEncoder(drop='first', handle_unknown='error'), ohe_features)
    ],
    remainder='passthrough'
)

# обучаем на train и трансформиуем ее же
X_train_lr = pd.DataFrame(
    col_transformer_lr.fit_transform(X_train_lr),
    columns = col_transformer_lr.get_feature_names_out()
)

# смотрим на преобразованные данные
X_train_lr.head()
````

````python
# трансформируем тестовую выборку
X_test_lr = pd.DataFrame(
    col_transformer_lr.transform(X_test_lr),
    columns=col_transformer_lr.get_feature_names_out()
)

# смотрим на преобразованные данные
X_test_lr.head()
````

Вот так будут выглядить данные после трансформации!

Но, учтем также и утечку данных, чтобы ее избежать воспользуемся конвеерами (Реализовано ниже в пункте 2.5)

### Разработка модели

Будем пробовать три модели:
1. *LogisticRegression*
2. *RandomForestClassifier*
3. *LGBMClassifier*

Линейная, "деревянная" и бустинг, посмотрим что даст наиболее внушительный результат.

#### LogisticRegression

````python
# вернем выборки в состояние до трансформации
X_train_lr = X_train.copy()
X_test_lr = X_test.copy()
````

````python
%%time

# создаем конвеер, используем col_transformer_lr, который мы определили выше
pipeline_lr = Pipeline([('transform', col_transformer_lr), ('clf', lr_model)])

# задаем сетку параметров
param_grid_lr = {
    'clf__penalty': ['l1', 'l2', 'elasticnet', 'none'],
    'clf__C': [0.1, 1, 10],
    'clf__solver': ['newton-cg', 'lbfgs', 'liblinear', 'sag', 'saga'],
    'clf__max_iter': [100, 500, 1000]
}

# описываем gs
gs_lr_pl = GridSearchCV(
    pipeline_lr, 
    param_grid=param_grid_lr, 
    scoring='roc_auc', 
    n_jobs=-1,
    cv=5
)

# фиттим модель
gs_lr_pl.fit(X_train_lr, y_train)

# сохраняем лучший скор и параметры
gs_lr_best_score = gs_lr_pl.best_score_
gs_lr_best_params = gs_lr_pl.best_params_


# выводим информацию на экран
print(f'LR BEST ROC AUC SCORE: {gs_lr_best_score:.2f}')
print(f'LR BEST PARAMS: {gs_lr_best_params}')
print('-' * 90)
````

Результат, мягко говоря, удручает. Попроубем использовать RandomForestClassifier.

Т.к. кодировать придется по другому (для RF будем использовать OrdinalEncoder), создадим отдельные train и test подвыборки

#### RandomForestClassifier

````python
# создаем подвыборку
X_train_rf = X_train.copy()
````

Кодировать будем похожим подходом

````python
# объявляем эстиматор
rf_model = RandomForestClassifier(
    random_state=RANDOM_STATE, 
    class_weight='balanced', 
    n_jobs=-1
)

# создаем трансформер
col_transformer_rf = ColumnTransformer(
    transformers=[
        ('ordinal', OrdinalEncoder(), ohe_features)
    ],
    remainder='passthrough'
)
````

````python
%%time

# создаем конвеер
pipeline_rf = Pipeline([('transform', col_transformer_rf), ('clf', rf_model)])

# задаем сетку параметров
param_grid_rf = {
    'clf__n_estimators': [230, 235, 240],
    'clf__max_depth': [6, 7, 8],
    'clf__min_samples_split': [10, 15, 20],
    'clf__min_samples_leaf': [1, 2, 3]
}

# описываем gs
gs_rf_pl = GridSearchCV(
    pipeline_rf, 
    param_grid=param_grid_rf, 
    scoring='roc_auc', 
    n_jobs=-1,
    cv=5
)

# фиттим модель
gs_rf_pl.fit(X_train_rf, y_train)

# сохраняем лучший скор и параметры
gs_rf_best_score = gs_rf_pl.best_score_
gs_rf_best_params = gs_rf_pl.best_params_


# выводим информацию на экран
print(f'RF BEST ROC AUC SCORE: {gs_rf_best_score:.2f}')
print(f'RF BEST PARAMS: {gs_rf_best_params}')
print('-' * 100)
````

Уже лучше, но пока недостаточно...

Пробуем модель с бустингом - LGBMClassifier

#### LGBMClassifier

````python
# создаем подвыборку
X_train_lgbm = X_train.copy()
````

````python
%%time

# объявляем эстиматор
lgbm_model = LGBMClassifier(
    random_state=RANDOM_STATE, 
    class_weight='balanced', 
    n_jobs=-1
)

# создаем конвеер
pipeline_lgbm = Pipeline([('transform', col_transformer_rf), ('clf', lgbm_model)])

# задаем сетку параметров
param_grid_lgbm = {
    'clf__n_estimators': range(280, 295, 1),
    'clf__max_depth': range(1, 40, 5),
    'clf__learning_rate': [0.01, 0.05, 0.1, 0.15, 0.2]
}

# описываем gs
gs_lgbm_pl = GridSearchCV(
    pipeline_lgbm, 
    param_grid=param_grid_lgbm, 
    scoring='roc_auc', 
    n_jobs=-1,
    cv=5
)

# фиттим модель
gs_lgbm_pl.fit(X_train_lgbm, y_train)

# сохраняем лучший скор и параметры
gs_lgbm_best_score = gs_lgbm_pl.best_score_
gs_lgbm_best_params = gs_lgbm_pl.best_params_


# выводим информацию на экран
print(f'LGBM BEST ROC AUC SCORE: {gs_lgbm_best_score:.2f}')
print(f'LGBM BEST PARAMS: {gs_lgbm_best_params}')
print('-' * 93)
````

Лучшей моделью себя показала LGBMClassifier: ROC AUC = 0.88

#### Тест лучшей модели

````python
%%time

# объявим эстиматор
best_model = gs_lgbm_pl.best_estimator_

# предсказываем значения
y_pred_proba = best_model.predict_proba(X_test)
y_pred = best_model.predict(X_test)

# считаем скоры
roc_auc = roc_auc_score(y_test, y_pred_proba[:,1])
f1 = f1_score(y_test, y_pred)
acc = accuracy_score(y_test, y_pred)

# выводим информацию на экран
print(f'BEST MODEL ROC AUC SCORE: {roc_auc:.2f}')
print(f'BEST MODEL F1-SCORE: {f1:.2f}')
print(f'BEST MODEL ACCURACY SCORE: {acc:.2f}')
print('-' * 31)
````

Лучшая модель дала довольно точные предсказания и ROC AUC и accuracy score больше 0.90, это неплохо, но можно сделать лучше.

#### График ROC CURVE

Значения метрик лучшей модели у нас есть, теперь построим график ROC-Кривой

````python
# сделаем размер поменьше
sns.set(rc={'figure.figsize':(10, 7)})

# получаем значения false positive rate, true positive rate и пороги
fpr, tpr, thresholds = roc_curve(y_test, y_pred_proba[:,1])

# строим ROC-Кривую
plt.plot(fpr, tpr, label='ROC-Кривая (area = 0.92)')
plt.plot([0, 1], [0, 1], 'k--', label='Случайное предположение')
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('ROC-Кривая')
plt.legend()
plt.show()
````

**Какой вывод можно сделать из графика ROC-кривой?**

ROC-Кривая показывает нам, как хорошо модель различает два класса в зависимости от выбранного порога вероятности. По оси y отложена доля правильно классифицированных положительных примеров относительно общего числа положительных примеров, а по оси x - доля неправильно классифицированных отрицательных примеров относительно общего числа отрицательных примеров.

*ROC-кривая имеет следующие свойства*:
- Чем ближе кривая к верхнему левому углу, тем лучше модель различает классы.
- Кривая, проходящая через точку (0,0) соответствует случайному предположению.
- Кривая, проходящая через точки (0,1) и (1,1), соответствует идеальной модели, которая всегда правильно классифицирует примеры.

ROC AUC - площадь под этой кривой, и наша метрика качества. После теста финальной модели получили значение 0.92, можем заключить, что модель получилась неплохой

````python
# Получение матрицы ошибок
tn, fp, fn, tp = confusion_matrix(y_test, y_pred).ravel()

# Создание pd.DataFrame для матрицы ошибок
data = {'Прогноз (+)': [tp, fp], 'Прогноз (-)': [fn, tn]}
index = ['Реальность (+)', 'Реальность (-)']
conf_matrix = pd.DataFrame(data=data, index=index)

# Вывод матрицы ошибок
print(conf_matrix)
````

Модель отлично справилась с классификацией негативного класса, это неудивительно, дисбаланс был сильно сдвинут в его сторону.

С положительным классом модель справилась неплохо.

````python
print(f'Доля правильно угаданных записей класса 0 = {tn / (tn + fp):.2f}')
print(f'Доля правильно угаданных записей класса 1 = {tp / (tp + fn):.2f}')
````

````python
# составные части матрицы ошибок
print("True Negative (TN):", tn)
print("False Positive (FP):", fp)
print("False Negative (FN):", fn)
print("True Positive (TP):", tp)

# метрики
precision = tp / (tp + fp)
recall = tp / (tp + fn)
f1 = 2 * (precision * recall) / (precision + recall)
accuracy = (tp + tn) / (tp + tn + fp + fn)

# вывод на экран
print('-' * 23)
print(f'Presicion = {precision:.2f}')
print(f'Recall = {recall:.2f}')
print(f'F1-Score = {f1:.2f}')
print(f'Accuracy = {accuracy:.2f}')
````
